# 特征

股价在n天内窄幅震荡，近期突破平台向上。

```python
from alpha.notebook import *
from alpha.core.rsi_stats import rsi30, rsiday

await init_notebook()
```

```python
def features(code, name, bars, results, frame_type):
    last = bars[-1]
    bars = bars[:-1]
    
    close, high, low, open_ = bars["close"], bars["high"], bars["low"], bars["open"]

    hhc = np.max(close)
    llc = np.min(close)
    am = (hhc - llc)/(hhc + llc)
    
    lclose = last['close']
    break_through = []
    
    for n in [20, 60, 100, 140]:
        break_through.append(lclose > np.max(close[-n:]))
        
    results.append([name, code, am, *break_through])
    
results = await scan(features, 141, nstocks=None)
df = pd.DataFrame(results, columns=["name", "code", "am", "b20", "b40", "b100", "b140"])
```
