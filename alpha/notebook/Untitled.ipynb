```python
def get_all_stocks(date: datetime.date)->List:
    secs = jq.get_all_securities(types=["stock"], date=end)
    
    is_st = jq.get_extras('is_st', secs.index.tolist(), start_date=date, end_date=date)
    
    no_st = []
    
    for sec in secs.index:
        if not is_st.iloc[0][sec]:
            no_st.append(sec)
            
    return no_st
```
